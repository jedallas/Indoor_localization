```python
file=pd.read_csv('C:/Users/user/Desktop/VIAX/content/13/cellid/process/wifiRssi.csv')
```

```python
for title in file.columns[1:]:
    model=file[title]
    model=model.dropna()
    list1=[]
    for i in model:
        list1.append(i)
    summary=0
    a=0.7
    b=1-a
    for i in range(0,len(list1)):
        if i==0:
            summary=list1[i]
        else:
            summary=a*summary+b*list1[i]
            list1[i]=summary
    index=file[title].isnull()
    j=0
    for i in range(0,len(file)):
        if index[i]==True:
            continue
        elif index[i]==False:
            file[title][i]=list1[j]
            j=j+1
    
```
